**Preparing the data for Model**

Scaling the data

```python
from sklearn.model_selection import train_test_split
```

```python
X= df.drop(['target'], axis=1)
y= df['target']
```

```python
X_train, X_test,y_train, y_test=train_test_split(X,y,test_size=0.3,random_state=40)
```

**Check the sample Size**

```python
print('X_train-', X_train.size)
print('X_test-',X_test.size)
print('y_train-', y_train.size)
print('y_test-', y_test.size)
```

**Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()

model1=lr.fit(X_train,y_train)
prediction1=model1.predict(X_test)
prediction1_train=model1.predict(X_train)
```

```python
X_test
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,prediction1)
```

```python
accuracy_score(y_train,prediction1_train)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, prediction1))
```

**Decision Tree**

```python
from sklearn.tree import DecisionTreeClassifier

dtc=DecisionTreeClassifier()
model2=dtc.fit(X_train,y_train)
prediction2=model2.predict(X_test)
prediction2_train=model2.predict(X_train)
```

```python
accuracy_score(y_test,prediction2)
```

```python
accuracy_score(y_train,prediction2_train)
```

```python
print(classification_report(y_test, prediction2))
```

**Random Forest**

```python
from sklearn.ensemble import RandomForestClassifier

rfc=RandomForestClassifier()
model3 = rfc.fit(X_train, y_train)
prediction3 = model3.predict(X_test)
prediction3_train= model3.predict(X_train)
```

```python
accuracy_score(y_test, prediction3)
```

```python
accuracy_score(y_train, prediction3_train)
```

```python
print(classification_report(y_test, prediction3))
```
